> Without this, your commits will have no name attached — GitHub will reject them.



````python
# ─────────────────────────────────────────────────────────────────────────────
#     Replace the name and email with your own details.
# ─────────────────────────────────────────────────────────────────────────────
#
# git config --global user.name  "Your Full Name"
# git config --global user.email "your_email@example.com"
# git config --global init.defaultBranch main
#
# ─── What each flag means ────────────────────────────────────────────────────
#   --global   →  applies to ALL repos on this machine (not just one project)
#   user.name  →  the name that appears on every commit you make
#   user.email →  must match your GitHub account email
#   init.defaultBranch main  →  ensures new repos start on 'main' (not 'master')
# ─────────────────────────────────────────────────────────────────────────────
````

## Section 1 — Python Version & Virtual Environment

A **virtual environment** is an isolated folder that holds one project's packages.  
It prevents version conflicts between different projects on your machine.


````python
# ─────────────────────────────────────────────────────────────────────────────
#  1.1  Check which Python is running
# ─────────────────────────────────────────────────────────────────────────────
import sys

print(f'Python executable : {sys.executable}')
print(f'Python version    : {sys.version}')

in_venv = ('.venv' in sys.executable or 'venv' in sys.executable
           or 'conda' in sys.executable)
print(f'Virtual env active? {"✅ YES" if in_venv else "❌ NO"}')

if not in_venv:
    print()
    print('  ⚠️  You are using the system Python.')
    print('  Run the commands in cell 1.2 to create and activate a venv first.')
````

## Section 5 — GitHub Workflow

| Term | Meaning |
|------|---------|
| `origin` | The remote copy of your repo on GitHub |
| `git clone` | Download a repo from GitHub to your machine |
| `git push` | Upload your local commits to GitHub |
| `git pull` | Download + merge GitHub changes into your local copy |
| Pull Request | Ask for your branch to be reviewed and merged into `main` |

````python
# ─────────────────────────────────────────────────────────────────────────────
# git remote add — connect your LOCAL repo to GitHub
#
#
#  STEP 1 — Create an empty repo on GitHub (github.com → New repository)
#
#  STEP 2 — Copy the HTTPS URL  (looks like):
#            https://github.com/<your-username>/<your-repo-name>.git
#
#  STEP 3 — Run in your terminal (replace the URL with yours):
#
#  git remote add origin https://github.com/<your-username>/<your-repo-name>.git
#
#  ── Breaking down the command ────────────────────────────────────────────────
#   git remote add   →  register a new remote connection
#   origin           →  the conventional nickname for your main GitHub remote
#                        (you can use any name, but 'origin' is the standard)
#   <url>            →  the full HTTPS (or SSH) address of your GitHub repo
#
#
#  STEP 4 — Push for the first time:
#
#  git push -u origin main
#
#  ── What -u does ─────────────────────────────────────────────────────────────
#   -u (short for --set-upstream) links your local 'main' branch to
#   'origin/main' on GitHub so that from now on a plain `git push` is enough.
# ─────────────────────────────────────────────────────────────────────────────
````

````python
# ─────────────────────────────────────────────────────────────────────────────
#  Key remote commands reference
#
# ── Link a local repo to GitHub (if you did git init locally) ────────────────
# git remote add origin https://github.com/<user>/repo.git
# git push -u origin main          # first push — sets the tracking branch
#
# ── Everyday push (after the first) ──────────────────────────────────────────
# git push
#
# ── Push a feature branch ────────────────────────────────────────────────────
# git push origin setup-task
#
# ── Get the latest from GitHub (teammates' commits) ──────────────────────────
# git pull origin main
#
````

## Section 6 — CI/CD with GitHub Actions

**CI (Continuous Integration)** = every `git push` automatically runs install + lint + tests.  
**CD (Continuous Deployment)** = if CI passes, automatically deploy the app.

```
git push
   │
   ▼
GitHub Actions triggered
   │
   ├─► 1. Checkout code
   ├─► 2. Set up Python 3.11
   ├─► 3. pip install -r requirements.txt
   ├─► 4. flake8  (lint)
   └─► 5. pytest  (tests)
              │
        ✅ PASS → safe to merge PR
        ❌ FAIL → fix before merging
```

The workflow file lives at: **`.github/workflows/ci.yml`**

````python
# ─────────────────────────────────────────────────────────────────────────────
#  Add CI workflow file 
#
# mkdir -p .github/workflows
# # ... create .github/workflows/ci.yml 
# git add .github/
# git commit -m "ci: add GitHub Actions workflow"
# ─────────────────────────────────────────────────────────────────────────────


ci_content = '''name: CI Pipeline

on:
  push:
    branches: [main]
  pull_request:
    branches: [main]

jobs:
  build-and-test:
    runs-on: ubuntu-latest
    steps:
      - uses: actions/checkout@v4
      - uses: actions/setup-python@v5
        with:
          python-version: "3.11"
      - run: python --version
      - run: pip install -r requirements.txt
      - run: flake8 src/ tests/ --max-line-length=88
      - run: pytest tests/ -v --tb=short
'''
````
